#### Joint PLOT
* It provides distribution for two variable or comparisons for 2 variables

| Plot Type    | Function        | Purpose                            |
| ------------ | --------------- | ---------------------------------- |
| Histogram    | `histplot()`    | Distribution of a variable         |
| KDE Plot     | `kdeplot()`     | Smooth distribution curve          |
| Bar Plot     | `barplot()`     | Compare means across categories    |
| Count Plot   | `countplot()`   | Frequency of categories            |
| Box Plot     | `boxplot()`     | Distribution with outliers         |
| Violin Plot  | `violinplot()`  | Box + KDE for distributions        |
| Scatter Plot | `scatterplot()` | Relationship between 2 variables   |
| Line Plot    | `lineplot()`    | Trend over sequence/time           |
| Heatmap      | `heatmap()`     | Correlation or matrix display      |
| Pairplot     | `pairplot()`    | All scatter plots between features |
| Jointplot    | `jointplot()`   | 2D scatter + hist + regression     |
| FacetGrid    | `FacetGrid()`   | Plot multiple charts for subgroups |


| `kind`      | Description                                                                 |
| ----------- | --------------------------------------------------------------------------- |
| `'scatter'` | Default; simple **scatter plot**.                                           |
| `'reg'`     | Scatter plot with a **regression line**.                                    |
| `'resid'`   | Shows **residuals** of a linear regression (useful for checking model fit). |
| `'kde'`     | **Kernel Density Estimation** plot (smoothed density).                      |
| `'hex'`     | **Hexbin plot** (good for large datasets – shows density using hexagons).   |
| `'hist'`    | **2D histogram** (alternative to hexbin when showing frequency).            |


```python
sns.jointplot(x='speeding',y='alcohol', data = crash_df , kind = 'hex')
```

#### KDE PLOT
* KDE (Kernel Density Estimation) is a smooth curve representing the probability density function of a continuous variable.
* It helps visualize the distribution of data more smoothly than a histogram.
* Think of it as a smoothed version of a histogram (like a mountain range over your data).

```python
sns.kdeplot(crash_df['alcohol'], color ='skyblue',fill=True)
```

```python
sns.histplot(crash_df['alcohol'], kde=True, color='blue')
plt.title("Histogram with KDE")
```

#### HEAT MAP

| Use Case                        | Example                                   |
| ------------------------------- | ----------------------------------------- |
| Correlation Matrix              | Visualize relationships between variables |
| Confusion Matrix in ML          | Show model performance                    |
| Pivot Tables / Aggregated Stats | Show frequency/averages across categories |
| Missing Data Map                | Visualize null values                     |


```python
plt.figure(figsize=(8,6))

crash_mx = crash_df.corr(numeric_only=True)
crash_mx
```

```python
ax = sns.heatmap(crash_mx, annot=True , fmt =".1f" , cmap ="magma", linecolor = "gray", annot_kws={"size": 9})
plt.title("Crash Correlation Heatmap", fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
colorbar = ax.collections[0].colorbar
colorbar.ax.tick_params(labelsize=10)  # change tick label size
colorbar.set_label("Correlation", fontsize=12)  # change label size/text
```

| Correlation Value | Meaning                      |
| ----------------- | ---------------------------- |
| `+1`              | Perfect positive correlation |
| `0`               | No correlation               |
| `-1`              | Perfect negative correlation |
| `+0.5 to +0.9`    | Strong positive              |
| `-0.5 to -0.9`    | Strong negative              |
| `+0.1 to +0.4`    | Weak positive                |
| `-0.1 to -0.4`    | Weak negative                |


```python
flights = sns.load_dataset('flights')
plt.figure(figsize =(12,8))
fl = flights.pivot_table(index = 'month' , columns = 'year', values = 'passengers', observed = False)
fl2 = fl.corr(numeric_only = True)
ax =sns.heatmap(fl2,cmap ='Blues', linecolor = 'white', linewidth = 1, annot = True ,annot_kws={"size": 9})
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
colorbar = ax.collections[0].colorbar
colorbar.ax.tick_params(labelsize=10)  # change tick label size
colorbar.set_label("Correlation", fontsize=12)  # change label size/text
```
