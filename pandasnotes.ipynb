##### PANDAS
1. pandas is a powerful library that helps to arrange data in a structured manner and also helps in data manipulation
2. Data structures are of three types series and dataframes and panles
3. Series: they are 1 dimensional arrays capable of holding any data type like sql

```python
df3 = pd.DataFrame({
    'Department': ['Sales', 'Sales', 'HR', 'HR', 'IT', 'IT'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [50000, 60000, 45000, 47000, 70000, 72000]
})
```

```python
df3
```

```python
# groupby
# syntax:
# df.groupby('Column name')[column].aggregate function
df3.groupby('Department')['Salary'].mean()
```

```python
df3.groupby(['Department','Employee'])['Salary'].sum()
```

```python
# multiple aggregate functions
df3.groupby('Department')['Salary'].agg(['mean','min','max'])
```

```python
# filtering
df3.groupby('Department')['Salary'].mean().reset_index().query('Salary > 60000')
```

```python
# Functions
df3['Increament']=df3['Salary'].apply(lambda x:x+500)
```
