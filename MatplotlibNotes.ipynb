#### TERMS TO KNOW FOR CREATING A GRAPH:

* fmt == shorthand notations [color][marker][line]
* label --> name of the line
* legend --> where the label should be placed or located that is legend(loc ='')
* color --> choose the color of the line or font or label
* marker --> marker that can be customized and this tracks the line meets
* markersize --> to define the size of the marker
* markeredgecolor --> the outline color of marker
* markerfacecolor --> the filling color of marker
* linestyle --> '--' , '-.' the style of line
* linewidth --> the width of line

```python
# creating plots
x = np.arange(0,5)
y = np.arange(0,10,2)

# plotting a line plot
plt.plot(x,y, label ='2X', color = 'red', linewidth = 2, marker = "o", markersize = 5 , markeredgecolor = "green" , linestyle ="--", markerfacecolor ="yellow")
plt.title('FIRST GRAPH !',fontdict ={'fontname':'Comic Sans MS' , 'fontsize':16,'color':'red'})
plt.xlabel("X- AXIS", fontdict ={'fontname':'Comic Sans MS'})
plt.ylabel("Y - AXIS",fontdict ={'fontname':'Comic Sans MS'})

# grid setup 
plt.grid(True, linestyle ='--',alpha =0.6, color = 'black')

# xticks --> x values and y ticks y values
plt.xticks([0,1,2,3,4,5])
plt.yticks([2,4,6,8,10])

# showing the labels and mentioning its position
plt.legend(loc='upper right')
plt.show()
```

#### BAR CHART 

| Situation                                               | Why Use a Bar Chart?                                                                          |
| ------------------------------------------------------- | --------------------------------------------------------------------------------------------- |
| 📊 You have **categorical data**                        | Bar charts clearly show comparisons across categories like countries, products, genders, etc. |
| 📈 You want to compare **discrete values**              | Great for comparing individual values (e.g., sales in Q1, Q2, Q3).                            |
| 📐 You want to display **frequency, counts, or totals** | Perfect for counts, percentages, or totals like survey responses or product sales.            |
| 👀 You want **clear, visual comparisons**               | Easily see which category is highest/lowest.                                                  |
| ⏱ You have **few categories**                           | Works best when the number of bars is small to moderate.                                      |


| Situation                                                | Why Not?                                                    |
| -------------------------------------------------------- | ----------------------------------------------------------- |
| You have **continuous data** (e.g., height, temperature) | Use a **histogram** instead.                                |
| You have **too many categories**                         | Use a **line chart** or **summary table** to avoid clutter. |
| You're showing **parts of a whole**                      | Consider a **pie chart** or **stacked bar** instead.        |


```python
# simple bar chart with labels and values 
# creating plots
labels = ['A','B','C']
values = [1,4,2]

# resize 
plt.figure(figsize =(5,3),dpi = 150)

# assigning the plt bar to a variable
# customizing bar chart
bars = plt.bar(labels, values,color = ['yellow' , 'green', 'red'], width = 0.5, edgecolor ='black', hatch = ['/','o','*']) # edge color = outline
# default width is 0.8

# using the variuable to set_hatch( That is pattern filling inside the bar) using for loop or normal way
## --> bars[0].set_hatch('/')
## --> bars[1].set_hatch('*')
## --> bars[2].set_hatch('o')

# adding values or labels inside the bars
# i --> gives the position of the bar 
# v/2 gives the values/2 that is the center of the bar
# syntax : plt.text(x[value or bar number] , y position , ha = horizontal alignmment m, va = vertical alignment , color = colr of the font , fontsize = text size of font)
for i , v in enumerate(values):
    plt.text(i, v/2, labels[i], ha= 'center', va='center', color='white', fontsize=14)


# alternative way :

"""
for i , bar in enumerate(bars):
    x = bar.get_x() + bar.get_width()/2  # x center of the bar
    y = bar.get_height()/2 
    plt.text(x,y,labels[i], color = 'white', fontsize = 16, ha ='center', va = 'center')
"""
# other ways to do the patterns that continuously print 
## --> patterns = ['*','/','o'] 
## --> for bar in bars:
     # this pops and returns the last item in the list
   ## -->  bar.set_hatch(patterns.pop())

plt.show()
```

#### HORIZONTAL BARS

```python
labels = ['A','B','C']
values = [2,4,2]
colors=['yellow','green','pink']

# altering figure size
# this should be done before plotting anything
plt.figure(figsize=(6,5),dpi = 150)


# creating bar
plt.barh(labels, values,label = colors, color=colors, hatch=['/', '*', 'o'])

# giving x and y labels with fontdict
plt.xlabel("Values", fontdict={'color': 'black', 'fontsize': 16, 'fontname': 'Comic Sans MS'})
plt.ylabel("Labels", fontdict={'color': 'black', 'fontsize': 16, 'fontname': 'Comic Sans MS'})

# giving label inside the bar
for i , v in enumerate(values):
    plt.text(v/2,i,labels[i],color ='white',ha='center', va='center',fontsize=16)
    
# adding grids
# alpha = 1.0 → fully opaque (no transparency)
# alpha = 0.0 → fully transparent (invisible)
# alpha = 0.6 → 60% visible, 40% transparent
plt.grid(axis='x', linestyle='--', alpha=0.6, color ='black')

# setting limits
plt.xlim(0, 8)

# adding title
plt.title('Horizontal Bar Chart' , fontdict = {'fontsize':18,'color':'black','fontname':'Comic Sans MS'})

# adding legend 
plt.legend(loc="upper right")

# saving the figure
plt.savefig('horibar.png',dpi =300)

# displaying the bar
plt.show()
```

#### stacked bar charts
* A stacked bar chart is used to show parts of a whole for each category, stacking multiple sub-categories in

```python
# 1. Vertical Stacked Bar Chart
```

```python
# giving plots
labels = ['A', 'B', 'C'] # base
group1 = [5, 7, 4] # catagory 1
group2 = [3, 2, 5] # category 2

# setting up figure size
plt.figure(figsize=(8,5))

# plotting the graph
plt.bar(labels, group1, label='Group 1', width=0.3,color=['skyblue','pink','yellow'], hatch='/')
plt.bar(labels, group2, bottom=group1,width=0.3,label='Group 2', color=['indigo','magenta','gold' ],hatch='o')

# setting x ticks and y ticks
plt.xticks(labels)
plt.xlabel("Categories")
plt.ylabel("Values")

# setting up grids
plt.grid(axis='y',alpha=0.6, linestyle='--',color='black')

#setting labels inside the graph
for i ,v in enumerate(group1):
    plt.text(i,v/2,str(v),color='white',ha='center',va='center',fontsize=12)
for i,v2 in enumerate(group2):
    plt.text(i,7.5,str(v2),color='white',ha='center',va='center',fontsize=12)
    
# setting title
plt.title("Stacked Bar Chart")
plt.legend(loc='upper left', fontsize=9, frameon=True, borderpad=0.5)
#-->frameon=True → adds a border box
#-->borderpad → adjusts space inside the legend box

#saving the figure
plt.savefig("stacked.png")

#clear displaying of the graph
plt.show()
```

#### PIE CHART

* A Pie Chart is a circular statistical chart that divides data into proportional slices.
* Each slice represents a category's contribution to the whole (100%).
  
##### basic syntax
> plt.pie(values, labels=None, colors=None, autopct=None, explode=None)

```python
# giving plots
labels = ['A', 'B', 'C']
sizes = [30, 45, 25]
colors = ['gold', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0)  # explode 2nd slice

# plotting the graph 
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.axis('equal')
plt.show()
```

#### REAL WORLD EXAMPLES

```python
# 1. Line graph

# loading data set and converting it to a data frame
gas_data = pd.read_csv('gas_prices.csv')
gas = pd.DataFrame(gas_data)

# altering figure
plt.figure(figsize=(8,5))

# creating a title
plt.title('Gas Prices Over Time (in USD)',fontdict={'fontname':'Comic Sans MS','color':'black','fontsize':14})

# creating plots
plt.plot(gas.Year[::3],gas.USA[::3],'b.-',label= 'United states')
plt.plot(gas.Year[::3], gas.Canada[::3],'c.-',label = 'Canada')
plt.plot(gas.Year[::3],gas['South Korea'][::3],'g.-',label = 'South Korea')
plt.plot(gas.Year[::3],gas.Italy[::3],'y.-',label='Italy')

# altering ticks
plt.xticks(gas.Year[::3])

# altering x and y labels
plt.xlabel('Year',fontdict={'fontname':'Comic Sans MS','color':'black','fontsize':12})
plt.ylabel('USD',fontdict={'fontname':'Comic Sans MS','color':'black','fontsize':12})
# legend
plt.legend(loc ='upper left',frameon = True,edgecolor='black', fontsize=9,borderpad=0.9)
plt.show()

```

```python
gas
countries_to_look_at =['Canada','France','Germany','Japan','Mexico','UK']

# creating graph for all the copuntries using for loop

for country in gas:
     if country in countries_to_look_at:
         plt.plot(gas.Year[::3],gas[country][::3],label=country,marker ='.',markerfacecolor='red',markeredgecolor='red')

# creating ticks
plt.xticks(gas.Year[::3])

# creating xlabels
plt.xlabel('Year')
plt.ylabel('USD')

# plt.title
plt.title("Gas Prices")

# legend
plt.legend(loc="upper left",frameon=True,borderpad=0.9,edgecolor="black")

plt.show()
```

```python
#### FIFA DATA

# data loading
data = pd.read_csv('fifa_data.csv')
fifa = pd.DataFrame(data)
fifa.head(5)
```

#### HISTOGRAM

```python
# plotting the ratings of a player
counts,bins, patches = plt.hist(fifa.Overall, bins=10, color='blue', edgecolor='black', alpha=0.7)

# title
plt.title("Distribution of Player Ratings")

# labels
plt.xlabel("Overall Rating")
plt.ylabel("Number of Players")

# adding labels ion the bins
for count, bin_edge, patch in zip(counts,bins,patches):
    x = patch.get_x() + patch.get_width() / 2
    y = patch.get_height()
    plt.text(x, y + 2, f"{int(round(count))}", ha='center', va='bottom', rotation=45, fontsize=10)
plt.show()
```

#### PIE CHART

```python
left = fifa[fifa['Preferred Foot']=='Left'].shape[0] # shape gives the number of rows
right = fifa[fifa['Preferred Foot']=='Right'].shape[0]

# style
plt.style.use('default')
# assigning labels
labels =['left','right']
# colors 
colors=['#abcdef','#aabbcc']
# explode
explode =[0,0.1]

# plotting the graph
plt.pie([left,right],labels = labels,colors = colors,explode =explode,autopct ='%1.1f%%')

# title
plt.title("Foot Preference Of FIFA Players",fontdict={"fontname":"Comic Sans MS","fontsize":12})
plt.show()
```

```python
# striping lbs from weight
fifa.Weight = [int(x.strip('lbs')) if type(x)== str else x for x in fifa.Weight]

# figure size
plt.figure(figsize = (6,14))

# Splitting values
light = fifa[fifa['Weight']<125].shape[0]
light_medium = fifa[(fifa['Weight']>=125) & (fifa['Weight']<150)].shape[0]
medium = fifa[(fifa['Weight']>=150) & (fifa['Weight']<175)].shape[0]
medium_heavy = fifa[(fifa['Weight']>=175) & (fifa['Weight']<200)].shape[0]
heavy = fifa[(fifa['Weight'] >=200)].shape[0]

# assigning values
weight = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125' , '125-150','150-175','175-200','Over 200']
color = ['#bc8ad1','#61042b','#0bd9d5','#dbd407','#8cf70a']
explode = [.4,.1,0,0,.4]
           
#plotting graph
plt.pie(weight , labels = labels, colors = color, autopct = '%.1f%%',pctdistance = 0.9, explode = explode )

# title
plt.title("Weight Distribution of FIFA Players (in lbs)")

plt.show()
```

#### BOX PLOT

```python
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
labels = ['barcelona','madrid', 'revs']

plt.figure(figsize = (5,8))

# plotting
plt.boxplot(
    [barcelona, madrid, revs],
    patch_artist=True,  # Required for facecolor to work
    tick_labels=['barcelona', 'madrid', 'revs'],  
    boxprops=dict(facecolor='lightblue', color='blue'),
    medianprops=dict(color='red'),
    meanprops=dict(marker='o', markerfacecolor='green'),
    showmeans=True
) 

plt.title("Professional Soccer Team Comparison")
plt.ylabel("FIFA overall rating")

plt.show()
```
