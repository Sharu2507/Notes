#### OUTLIERS
* Outlier is an extreme or odd value in the data
* A quantile splits your data into parts.

Think of quantiles as percent-based dividers:

25% quantile (Q1) = the value below which 25% of the data falls

50% quantile (Q2) = the median (middle value)

75% quantile (Q3) = the value below which 75% of the data falls

#### METHODS TO CLEAN OUTLIERS
* There are 3 main methods to clean outliers
    
| Method                | Use Case                                                  | Easy to Use? |
| --------------------- | --------------------------------------------------------- | ------------ |
| **1. IQR Method**     | Most common for basic data cleaning                       | ✅ Yes        |
| **2. Z-score Method** | Good for normally distributed data (bell curve)           | ⚠️ Medium    |
| **3. Visualization**  | Helps **see** outliers using plots (boxplot, scatterplot) | ✅ Yes        |


##### IQR (INTER QUARTILE RANGE)
* It looks at the middle 50% of the data and flags anything too far outside as an outlier.
* Sort the column values
* Find:
1. Q1 = 25% of data is below this point
2. Q3 = 75% of data is below this point
>Compute:
* IQR = Q3 - Q1
* Mark any value as an outlier if:
* It is less than Q1 -1.5×IQR or greater than Q3 + 1.5×IQR


##### 📊 What Is a Quantile?
* A **quantile** splits your data into parts.
* Think of quantiles as **percent-based dividers**:

1. **25% quantile (Q1)** = the value below which **25% of the data** falls
2. **50% quantile (Q2)** = the **median** (middle value)
3. **75% quantile (Q3)** = the value below which **75% of the data** falls

###### 🎯 Example:
* Imagine these 8 numbers (sorted):
```
[10, 20, 25, 30, 35, 40, 45, 50]
```
* **Q1 (25% quantile)** = 22.5 → Between 20 and 25
* **Q2 (50%)** = 32.5 → Middle of 30 and 35
* **Q3 (75%)** = 42.5 → Between 40 and 45

These points split the data into **4 equal parts**.

###### 📐 Why Is It Useful?
    
We use **Q1 and Q3** to measure the **spread** of the middle data.
The range between them is called:
### 🔸 **IQR = Q3 - Q1**

It helps us **ignore extreme values (outliers)** and focus on the central values.

---

### 🧪 In Pandas:

```python
df['LIMIT_BAL'].quantile(0.25)  # Q1
df['LIMIT_BAL'].quantile(0.75)  # Q3
```

This finds the values that separate the **lowest 25%** and **highest 25%**.

---

### ✅ Recap:

| Quantile | Symbol | Meaning                         |
| -------- | ------ | ------------------------------- |
| 25%      | Q1     | 25% of data is below this point |
| 50%      | Q2     | Median — the middle of the data |
| 75%      | Q3     | 75% of data is below this point |

---

##### Z - SCORE

* It checks how far a value is from the mean, using standard deviation.
* Find the mean and standard deviation of the column
> Use formula:
* Z= (x−mean) /std
* If Z > 3 or Z < -3, the value is an outlier.

##### Visualization Methods — To See Outliers
1. Boxplot:
* The box shows the middle 50% of the data
* Dots outside the box are outliers
2. Scatter plot:
* Good for spotting outliers in 2D (like AGE vs Salary)

#### STEP BY STEP CALCULATION OF OUTLIERS

````python
# 1.  Find Q1 and Q3
df.columns
````

````python
q1 = df['AGE'].quantile(0.25)
q3 = df['AGE'].quantile(0.75)
print("Q1 :",q1)
print("Q3 :", q3)
````

````python
# 2. calculate the IQR 
IQR = q3 - q1
print("IQR :", IQR)
````

````python
# 3. Calculate Outlier Limits
lower_limit = q1 - 1.5*IQR
upper_limit = q3+1.5*IQR
print("LOWER LIMIT :" ,lower_limit)
print("UPPER LIMIT :" , upper_limit)
````

````python
outliers = df[(df['AGE'] < lower_limit) | (df['AGE'] > upper_limit)]
print(outliers)
````

````python
print(f"Number of outliers: {outliers.shape[0]}")
````

````python
#### CLEANINIG ALL THE OUTLIERS
num_cols = df.select_dtypes(include =['int64', 'float64']).columns
````

````python
for col in num_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    IQR = q3 - q1
    low_limit = q1 -1.5*IQR
    up_limit = q3+1.5*IQR
    df[col] = df[col].clip(low_limit,up_limit)
````
